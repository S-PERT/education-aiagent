## 기억력 만들기

```python
import openai

client = openai.Client()

history = []

def call_ai():
	response = client.chat.completions.create(
		model="gpt-4o-mini",
		messages=history
	)
	assistant_message = response.choices[0].message.content
	history.append({"role": "assistant", "content": assistant_message})    	
	print("Assistant:", assistant_message)

while True:
	user_input = input("User: ")
	if user_input.lower() in ["exit", "quit", "q"]:
		print("Exiting the chat.")
		break
	else:
		print("User:", user_input)        
		history.append({"role": "user", "content": user_input})
		response = call_ai()
```

## 도구 알려주기

```python
import openai

client = openai.Client()

history = []

TOOLS = [
	{
		"type": "function",
		"function": {
			"name": "get_weather",
			"description": "Get the current weather for a given country.",
			"parameters": {
				"type": "object",
				"properties": {
					"country": {
						"type": "string",
						"description": "The name of the country to get the weather for.",
					}
				},
				"required": ["country"]
			}
		}
	}
]

def call_ai():
	response = client.chat.completions.create(
		model="gpt-4o-mini",
		messages=history,
		tools=TOOLS,
	)
	print("DEBUG: Full response:", response)
	assistant_message = response.choices[0].message.content
	history.append({"role": "assistant", "content": assistant_message})    	
	print("Assistant:", assistant_message)

while True:
	user_input = input("User: ")
	if user_input.lower() in ["exit", "quit", "q"]:
		print("Exiting the chat.")
		break
	else:
		print("User:", user_input)        
		history.append({"role": "user", "content": user_input})
		response = call_ai()
```

## 도구 사용하기

```python
import openai, json
import openai.types.chat as ChatCompletionMessage

def get_weather(country): 
	return "낮술 먹기 딱 좋은 날씨입니다."

FUNTION_MAP = { "get_weather": get_weather }


client = openai.Client()

history = []

TOOLS = [
	{
		"type": "function",
		"function": {
			"name": "get_weather",
			"description": "Get the current weather for a given country.",
			"parameters": {
				"type": "object",
				"properties": {
					"country": {
						"type": "string",
						"description": "The name of the country to get the weather for.",
					}
				},
				"required": ["country"]
			}
		}
	}
]

def response_processing(message: ChatCompletionMessage): 
	requestTools = message.tool_calls 							
	if requestTools: 											 ## 응답에 도구 사용이 있는지 체크
		id = requestTools[0].id 					
		function_name = requestTools[0].function.name 
		arguments = requestTools[0].function.arguments 
		print(f"AI requested tool: {function_name} with arguments {arguments}")  
		
		history.append(
			{ 
				"role": "assistant", 
				"content": message.content or "", 
				"tool_calls": [
					{ 
						"id": id, 
						"type": "function", 
						"function": { 
							"name": function_name, 
							"arguments": arguments 
						} 
					}
				]
			}
		) 
		try: 
			arguments = json.loads(arguments) 							 ## 호출 인자 JSON 파싱
		except json.JSONDecodeError as e: 
			arguments = {} 

		if function_name in FUNTION_MAP: 
			function_response = FUNTION_MAP[function_name](**arguments) 		 ## 함수 호출
			history.append({"role": "tool", "tool_call_id": id, 
				"name": function_name, "content": function_response}) 
			
		call_ai() 
	else: 
		assistant_message = message.content 
		history.append({"role": "assistant", "content": assistant_message}) 
		print("Assistant:", assistant_message)


def call_ai():
	response = client.chat.completions.create(
		model="gpt-4o-mini",
		messages=history,
		tools=TOOLS,
	)
	response_processing(response.choices[0].message)
	
	# assistant_message = response.choices[0].message.content
	# history.append({"role": "assistant", "content": assistant_message})    	
	# print("Assistant:", assistant_message)

while True:
	user_input = input("User: ")
	if user_input.lower() in ["exit", "quit", "q"]:
		print("Exiting the chat.")
		break
	else:
		print("User:", user_input)        
		history.append({"role": "user", "content": user_input})
		response = call_ai()
```
